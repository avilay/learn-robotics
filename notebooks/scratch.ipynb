```python
def normalized(raw):
    range_max, range_min = 3211, 862
    mid = (range_max + range_min) / 2
    return (raw - mid) * 360 / 4095
```

```python
def raw(normalized):
    range_max, range_min = 3211, 862
    mid = (range_max + range_min) / 2
    return (normalized * 4095 / 360) + mid
```
