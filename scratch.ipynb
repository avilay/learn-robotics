```python
def build_lerp(v0, vT, T):
    def lerp(t):
        return (vT - v0) * t / T + v0

    return lerp
```

```python
lerp = build_lerp(1, 10, 5)
```

```python
for t in range(1, 6):
    print(lerp(t))
```

```python
def build_lerp(init, final, tot_ticks):
    def lerp(tick):
        return (final - init) * tick / tot_ticks + init

    return lerp
```

```python
lerp = build_lerp(1, 10, 5)
```

```python
for t in range(1, 6):
    print(lerp(t))
```
